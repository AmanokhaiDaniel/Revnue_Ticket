**Project Description**

This project involves a detailed analysis of revenue trends for key companies over a specific period (December 2023 to June 2024). The focus is on examining the revenue performance of companies like BOLT, CAR HIRE, and TASK FORCE. The project includes data extraction, visualization, and the generation of insights through a structured report.

```python
!pip install python-docx
import pandas as pd

# Load the Excel file for the data analysis
file_path = '/content/REVENUE CHECK DEC 2023-JUNE 2024 (1).xlsx'
xls = pd.ExcelFile(file_path)

# Extract sheet names to determine which sheets to load
sheet_names = xls.sheet_names
sheet_names

# Load the Monthly Comparative Analysis and Summary sheets for analysis
monthly_comparative = pd.read_excel(xls, sheet_name='MONTHLY COMPARATIVE ANALYSIS')
summary = pd.read_excel(xls, sheet_name='SUMMARY')

# Display the first few rows of the monthly comparative analysis and summary sheets
monthly_comparative_head = monthly_comparative.head()
summary_head = summary.head()

monthly_comparative_head, summary_head

import matplotlib.pyplot as plt

# Plotting the Monthly Revenue for BOLT, CAR HIRE, and TASK FORCE
plt.figure(figsize=(10, 6))

# Plotting each category
plt.plot(monthly_comparative['MONTHS'], monthly_comparative['BOLT'], label='BOLT', marker='o')
plt.plot(monthly_comparative['MONTHS'], monthly_comparative['CAR HIRE'], label='CAR HIRE', marker='o')
plt.plot(monthly_comparative['MONTHS'], monthly_comparative['TASK FORCE'], label='TASK FORCE', marker='o')

# Adding labels and title
plt.title('Monthly Revenue Analysis (BOLT AND CAR HIRE,)', fontsize=14)
plt.xlabel('Months', fontsize=12)
plt.ylabel('Revenue (NGN)', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the plot as an image to include in the document
plt_path = "monthly_revenue_analysis.png"
plt.savefig(plt_path)

plt.show()
from docx import Document
from docx.shared import Inches

# Create a new Document
doc = Document()
doc.add_heading('Revenue Analysis and Inference Report', 0)

# Add introduction
doc.add_paragraph(
    "This report provides a detailed analysis of the revenue trends for the period of December 2023 to June 2024. "
    "The focus is on key companies such as BOLT, CAR HIRE, and TASK FORCE, with a special emphasis on monthly revenue growth."
)

# Add a section for Monthly Comparative Analysis
doc.add_heading('Monthly Comparative Analysis', level=1)

# Insert the chart image
doc.add_paragraph('The following chart illustrates the revenue trends for BOLT, CAR HIRE, and TASK FORCE:')
doc.add_picture(plt_path, width=Inches(6))

# Add more detailed analysis and insights
doc.add_paragraph(
    "From the chart, we can observe the following key trends:\n"
    "- BOLT has consistently maintained a strong revenue performance, showing steady growth over the months.\n"
    "- CAR HIRE and TASK FORCE also show positive growth, but at a lower rate compared to BOLT.\n"
    "- The percentage growth in revenue indicates an overall upward trend, suggesting a healthy increase in market demand."
)

# Add a section for Inferences
doc.add_heading('Inferences', level=1)
doc.add_paragraph(
    "Based on the data and analysis, the following inferences can be made:\n"
    "1. **Dominance of BOLT:** BOLT continues to dominate the market with the highest revenue figures, making it a market leader.\n"
    "2. **Growth Opportunities:** CAR HIRE and TASK FORCE have potential for growth, particularly if they can capitalize on the increasing market demand.\n"
    "3. **Digital Adoption:** Encouraging digital transactions could further enhance revenue, especially for companies like TASK FORCE that show room for improvement."
)

# Add a conclusion
doc.add_heading('Conclusion', level=1)
doc.add_paragraph(
    "The analysis indicates strong market performance for the key companies. Continued focus on revenue growth strategies, "
    "especially digital adoption, will be essential for maintaining and enhancing market positions."
)

# Save the document
doc_path = "Revenue_Analysis_Report.docx"
doc.save(doc_path)

doc_path


```
